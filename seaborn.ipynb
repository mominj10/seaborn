```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
titanic=sns.load_dataset('titanic')
planets=sns.load_datset('planets')
```

```python
tips
```

```python
sns.scatterplot(x='tip',y='total_bill',data=tips)
```

```python
#add hue
sns.scatterplot(x='tip',y='total_bill',data=tips,hue='day')
```

```python
#add hue
sns.scatterplot(x='tip',y='total_bill',data=tips,hue='day',size='size')
```

```python
#histogram plot
#kde=True: This parameter adds a kernel density estimate to the histogram, providing a smoothed line that estimates the underlying distribution of the data.
#bins=15: This parameter specifies the number of bins or intervals in the histogram. In this case, it's set to 15.
sns.histplot(tips['tip'],kde=True,bins=15)
```

```python
sns.displot(tips['tip'],kde=True,bins=15)
```

```python
#barplot
import matplotlib.pyplot as plt
sns.barplot(x='sex',y='tip',data=tips,palette='Blues')
```

```python
#boxplot
sns.boxplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.stripplot(x='day',y='tip',data=tips,hue='sex',dodge=True)
#dodge seperates the values, in this case male and female
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='reg')
#kind='reg': Specifies the type of plot. In this case, 'reg' stands for regression, so it will include a regression line.
#kind='scatter': Default option, creates a scatter plot.
#kind='reg': Adds a regression line and a scatter plot.
#kind='hex': Creates a hexbin plot.
#kind='kde': Adds kernel density estimates along with the scatter plot.
#kind='resid': Displays the residuals of the regression on the y-axis.
#kind='hist': Adds histograms on the marginal axes.
```

```python
#adds kernel density
sns.jointplot(x='tip',y='total_bill',data=tips,kind='kde',shade=True)
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='hex')
```
